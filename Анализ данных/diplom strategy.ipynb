```python
def get_signal (tick):
    i = tick.shift(+1).join(beta1)
    i = i.dropna()

    X = sm.add_constant(i[['Topic_738', 'Topic_1076', 'Topic_6',
                                  'Topic_854', 'Topic_668', 'Topic_722', 
                                  'Topic_1293', 'Topic_1122', 'Topic_11',
                                  'Topic_183', 'Topic_817', 'Topic_1269', 
                                  'Topic_873', 'Topic_749', 'Topic_1012', 
                                  'Topic_512', 'Topic_2003', 'Topic_333', 
                                  'Topic_1003', 'Topic_876', 'Topic_20', 
                                  'Topic_1269', 'Topic_97', 'Topic_21']].loc[:'2018-01-10'])
    y = i['Close'].loc[:'2018-01-10']
    model = sm.OLS(y, X)
    results = model.fit()
    
    return("Negative", results.params[(results.pvalues<0.05) & (results.params<0)].index.tolist(),
          "Positive", results.params[(results.pvalues<0.05) & (results.params>0)].index.tolist())
    #print( "Positive", results.params[(results.pvalues<0.05) & (results.params>0)].index)
```
