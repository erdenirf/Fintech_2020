```python
import pandas as pd
import os
os.chdir("/Users/iakubovskii/Machine_Learning/Datasets/MovieLens/ml-25m")
movies = pd.read_csv('movies.csv', engine='python')
genome_scores = pd.read_csv("genome-scores.csv")
genome_tags = pd.read_csv("genome-tags.csv")
genome_scores.pivot_table(index = 'movieId', 
                          columns = 'tagId',
                          values = 'relevance')

import re
import numpy as np
def extract_year(title):
    try: 
        year = round(int(re.search(r"\d{4}", title).group(0)),0)
    except:
        year = np.nan
    return(year)
movies['year'] = movies['title'].map(extract_year)
movies['year'] = pd.array(movies['year'].values, dtype = pd.Int64Dtype()) 
movies[(movies.year < 2025) & (movies.year > 1994)]['year'].plot(kind = 'hist');
```
